since our data is is clean now we check for outliers.



A data point that deviates significantly from the majority of other observations in a dataset, either by being much higher or much lower in value, or by differing in context or collective behavior is known as OUTLIER.

```python
fig, axs = plt.subplots(3,3, figsize = (20,15))
plt1 = sns.boxplot(housing['price'], ax = axs[0,0])
plt2 = sns.boxplot(housing['floors'], ax = axs[0,1])
plt3 = sns.boxplot(housing['bedrooms'], ax = axs[0,2])
plt1 = sns.boxplot(housing['bathrooms'], ax = axs[1,0])
plt2 = sns.boxplot(housing['condition'], ax = axs[1,1])
plt3 = sns.boxplot(housing['yr_built'], ax = axs[1,2])
plt1 = sns.boxplot(housing['country'], ax = axs[2,0])
plt2 = sns.boxplot(housing['sqft_living'], ax = axs[2,1])
plt3 = sns.boxplot(housing['yr_renovated'], ax = axs[2,2])

plt.tight_layout()
```

clearly, 'bathrooms', 'sqft_living','yr_living' , and 'floors'  has considerable outlier.
now to fix this, we perform:

```python
plt.boxplot(housing.price)
Q1 = housing.bathrooms.quantile(0.25)
Q3 = housing.bathrooms.quantile(0.75)
IQR = Q3 - Q1
housing = housing[(housing.bathrooms >= Q1 - 1.5*IQR) & (housing.bathrooms <= Q3 + 1.5*IQR)]
```

```python
plt.boxplot(housing.price)
Q1 = housing.floors.quantile(0.25)
Q3 = housing.floors.quantile(0.75)
IQR = Q3 - Q1
housing = housing[(housing.floors >= Q1 - 1.5*IQR) & (housing.floors <= Q3 + 1.5*IQR)]
```

```python
plt.boxplot(housing.price)
Q1 = housing.sqft_living.quantile(0.25)
Q3 = housing.sqft_living.quantile(0.75)
IQR = Q3 - Q1
housing = housing[(housing.sqft_living >= Q1 - 1.5*IQR) & (housing.sqft_living <= Q3 + 1.5*IQR)]
```

```python
plt.boxplot(housing.price)
Q1 = housing.price.quantile(0.25)
Q3 = housing.price.quantile(0.75)
IQR = Q3 - Q1
housing = housing[(housing.price >= Q1 - 1.5*IQR) & (housing.price <= Q3 + 1.5*IQR)]
```
