```python
plt.boxplot(housing.price)
Q1 = housing.price.quantile(0.25)
Q3 = housing.price.quantile(0.75)
IQR = Q3 - Q1
housing = housing[(housing.price >= Q1 - 1.5*IQR) & (housing.price <= Q3 + 1.5*IQR)]
```

```python
def univariate_analysis_numeric(column):
    plt.figure(figsize=(15, 5)) # Set figure size here

    # Histogram
    plt.subplot(1, 3, 1) # Create 1 row and 3 columns, select the 1st plot
    sns.histplot(housing[column], kde=True)
    plt.title(f'Distribution of {column}')

    # Box plot
    plt.subplot(1, 3, 2) # Select the 2nd plot
    sns.boxplot(x=housing[column])
    plt.title(f'Box Plot of {column}')

    # Violin plot
    plt.subplot(1, 3, 3) # Select the 3rd plot
    sns.violinplot(x=housing[column])
    plt.title(f'Violin Plot of {column}')

    plt.tight_layout()
    plt.show()
```

Breakdown of the function

* plt.figure(figsize=(10, 4)): This sets up the plotting environment, creating a figure with a specified width and height to ensure the plots are clearly displayed side-by-side.
* plt.subplot(1, 2, 1): This indicates that the next plot will be placed in the first position of a grid with one row and two columns.
* sns.histplot(housing[column], kde=True): This creates a histogram of the specified numeric column using the Seaborn library.
* Histogram: A bar chart that shows the frequency distribution of a continuous variable. It groups data into "bins" and displays the number of data points in each bin, allowing you to see the overall shape of the data.
* kde=True: This overlay adds a Kernel Density Estimate (KDE) curve, which is a smoothed, continuous representation of the data's distribution.
* plt.title(f'Distribution of {column}'): This sets the title for the histogram plot.
* plt.subplot(1, 2, 2): This switches to the second position in the plot grid for the next graph.
* sns.boxplot(x=housing[column]): This creates a box plot of the specified numeric column using Seaborn.
* Box plot: A standardized way to display the distribution of numerical data using a "five-number summary". The box itself shows the interquartile range (IQR), while the central line marks the median.
* plt.title(f'Box Plot of {column}'): This sets the title for the box plot.
* plt.tight_layout(): This automatically adjusts the plot parameters to give a tight layout, preventing titles and labels from overlapping.
* plt.show(): This displays the generated plots.

**What these plots represent**
By visualizing a single numerical column in these two ways, you gain a comprehensive understanding of its properties, which is crucial for exploratory data analysis (EDA).

Histogram interpretation:
* Shape: Is the distribution symmetrical (bell-shaped), skewed to the left or right, or uniform?
* Central tendency: Where is the peak of the data? This indicates the most frequent values.
* Spread: How wide is the distribution? A wide spread indicates high variability.
* Modes: Does the distribution have one peak (unimodal) or more (bimodal or multimodal)?

Box plot interpretation:
* Outliers: The individual points beyond the "whiskers" show potential outliers that are statistically different from the rest of the data.
* Symmetry and skew: You can easily assess the skewness of the data. If the median is not in the middle of the box, or if one whisker is much longer, the data is skewed.
* Range and IQR: The length of the whiskers and the box itself shows the overall range and the spread of the middle 50% of the data, respectively.

**What a violin plot indicate**

Key elements and their meaning
* the shape of the violin: The overall shape of the plot represents the kernel density estimation (KDE), which shows the probability density of the data at different values.
* Wider sections: Indicate a higher concentration of data points.
* Thinner sections: Indicate a lower concentration of data points.
